## Створюємо функцію

```python
def task_6():
    numbers = input()
    a, b = map(int, numbers.split())
    
    if a > b:
        print("Початок діапазону не може бути більшим за кінець")
        return[]

    primes = []
    for i in range (a, b + 1):
        if i <= 1:
            continue
        
        is_prime = True
        for j in range(2, int(i ** 0.5) + 1):
            if i % j == 0:
                is_prime = False
                break

        if is_prime:
            primes.append(i)
    return primes
```
